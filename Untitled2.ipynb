```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cvx

X1 = np.array([[-1.1,0],[-0.3,0.1],[-0.9,1],[0.8,0.4],[0.4,0.9],[0.3,-0.6],
               [-0.5,0.3],[-0.8,0.6],[-0.5,-0.5]])                                   # 9개 class1 data

X0 = np.array([[-1,-1.3], [-1.6,2.2],[0.9,-0.7],[1.6,0.5],[1.8,-1.1],[1.6,1.6],
               [-1.6,-1.7],[-1.4,1.8],[1.6,-0.9],[0,-1.6],[0.3,1.7],[-1.6,0],[-2.1,0.2]])    # 13개 class0 data

X1 = np.asmatrix(X1)
X0 = np.asmatrix(X0)

plt.figure(figsize = (6, 4))
plt.plot(X1[:,0], X1[:,1], 'ro', markersize = 4, label = 'Class 1')
plt.plot(X0[:,0], X0[:,1], 'bo', markersize = 4, label = 'Class 0')
plt.title('Logistic Regression for Nonlinear Data')
plt.xlabel(r'x')
plt.ylabel(r'y')
plt.legend(loc = 4)
plt.show()
```

```python
N = X1.shape[0]
M = X0.shape[0]

X = np.vstack([X1, X0])
y = np.vstack([np.ones([N,1]), -np.ones([M,1])])  # 1 : N개, -1 : M개

X = np.asmatrix(X)
y = np.asmatrix(y)

m = N + M
Z = np.hstack([np.ones([m,1]), np.square(X[:,0]), np.sqrt(2)*np.multiply(X[:,0],X[:,1]), np.square(X[:,1])])

w = cvx.Variable([4, 1])

obj = cvx.Minimize(cvx.sum(cvx.logistic(-cvx.multiply(y,Z @ w))))   #  cvxpy.multiply(lh_expr, rh_expr)
prob = cvx.Problem(obj).solve()

w = w.value
print(w)
```

```python
# to plot

[X1gr, X2gr] = np.meshgrid(np.arange(-3,3,0.1), np.arange(-3,3,0.1)) # 둘다 60 x 60

Xp = np.hstack([X1gr.reshape(-1,1), X2gr.reshape(-1,1)])
Xp = np.asmatrix(Xp)

m = Xp.shape[0]
Zp = np.hstack([np.ones([m,1]), np.square(Xp[:,0]), np.sqrt(2)*np.multiply(Xp[:,0], Xp[:,1]), np.square(Xp[:,1])])
q = Zp*w

B = []
for i in range(m):
    if q[i,0] > 0:
        B.append(Xp[i,:])

B = np.vstack(B)

plt.figure(figsize = (6, 4))
plt.plot(X1[:,0], X1[:,1], 'ro', markersize = 4, label = 'Class 1')
plt.plot(X0[:,0], X0[:,1], 'bo', markersize = 4, label = 'Class 0')
plt.plot(B[:,0], B[:,1], 'ys', markersize = 10, alpha = 0.1, label = 'Logistic Regression')
plt.title('Logistic Regression with Kernel')
plt.xlabel(r'x')
plt.ylabel(r'y')
plt.legend(loc = 1)
plt.show()
     
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cvx

X1 = np.array([[-1.1,0],[-0.3,0.1],[-0.9,1],[0.8,0.4],[0.4,0.9],[0.3,-0.6],
               [-0.5,0.3],[-0.8,0.6],[-0.5,-0.5]])                                   # 9개 class1 data

X0 = np.array([[-1,-1.3], [-1.6,2.2],[0.9,-0.7],[1.6,0.5],[1.8,-1.1],[1.6,1.6],
               [-1.6,-1.7],[-1.4,1.8],[1.6,-0.9],[0,-1.6],[0.3,1.7],[-1.6,0],[-2.1,0.2]])    # 13개 class0 data

X1 = np.asmatrix(X1)
X0 = np.asmatrix(X0)

plt.figure(figsize = (6, 4))
plt.plot(X1[:,0], X1[:,1], 'ro', markersize = 4, label = 'Class 1')
plt.plot(X0[:,0], X0[:,1], 'bo', markersize = 4, label = 'Class 0')
plt.title('Logistic Regression for Nonlinear Data')
plt.xlabel(r'x')
plt.ylabel(r'y')
plt.legend(loc = 4)
plt.show()
```

```python
N = X1.shape[0]
M = X0.shape[0]

X = np.vstack([X1, X0])
y = np.vstack([np.ones([N,1]), -np.ones([M,1])])  # 1 : N개, -1 : M개

X = np.asmatrix(X)
y = np.asmatrix(y)

m = N + M
Z = np.hstack([np.ones([m,1]),np.sqrt(2)*(X[:,0]),np.sqrt(2)*(X[:,1]), np.square(X[:,0]), np.sqrt(2)*np.multiply(X[:,0],X[:,1]), np.square(X[:,1])])

w = cvx.Variable([6, 1])

obj = cvx.Minimize(cvx.sum(cvx.logistic(-cvx.multiply(y,Z @ w))))   #  cvxpy.multiply(lh_expr, rh_expr)
prob = cvx.Problem(obj).solve()

w = w.value
print(w)
```

```python
# to plot

[X1gr, X2gr] = np.meshgrid(np.arange(-3,3,0.1), np.arange(-3,3,0.1)) # 둘다 60 x 60

Xp = np.hstack([X1gr.reshape(-1,1), X2gr.reshape(-1,1)])
Xp = np.asmatrix(Xp)

m = Xp.shape[0]
Zp = np.hstack([np.ones([m,1]),np.sqrt(2)*(Xp[:,0]),np.sqrt(2)*(Xp[:,1]), np.square(Xp[:,0]), np.sqrt(2)*np.multiply(Xp[:,0], Xp[:,1]), np.square(Xp[:,1])])
q = Zp*w

B = []
for i in range(m):
    if q[i,0] > 0:
        B.append(Xp[i,:])

B = np.vstack(B)

plt.figure(figsize = (6, 4))
plt.plot(X1[:,0], X1[:,1], 'ro', markersize = 4, label = 'Class 1')
plt.plot(X0[:,0], X0[:,1], 'bo', markersize = 4, label = 'Class 0')
plt.plot(B[:,0], B[:,1], 'ys', markersize = 10, alpha = 0.1, label = 'Logistic Regression')
plt.title('Logistic Regression with Kernel')
plt.xlabel(r'x')
plt.ylabel(r'y')
plt.legend(loc = 1)
plt.show()
     
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline

# 3 class의 cluster data 생성
mu1 = np.array([1, 7])  # mean value
mu2 = np.array([3, 4])
mu3 = np.array([6, 5])

SIGMA1 = 0.8*np.array([[1, 1.5],
                       [1.5, 3]])
SIGMA2 = 0.5*np.array([[2, 0],
                       [0, 2]])
SIGMA3 = 0.5*np.array([[1, -1],
                       [-1, 2]])

X1 = np.random.multivariate_normal(mu1, SIGMA1, 100)
X2 = np.random.multivariate_normal(mu2, SIGMA2, 100)
X3 = np.random.multivariate_normal(mu3, SIGMA3, 100)
y1 = 1*np.ones([100,1])
y2 = 2*np.ones([100,1])
y3 = 3*np.ones([100,1])

plt.figure(figsize = (6, 4))
plt.title('Generated Data for Multiclass Classification', fontsize=14)
plt.plot(X1[:,0], X1[:,1], '.', markersize = 4, label = 'Class 1')
plt.plot(X2[:,0], X2[:,1], '.', markersize = 4, label = 'Class 2')
plt.plot(X3[:,0], X3[:,1], '.', markersize = 4, label = 'Class 3')
plt.xlabel('x', fontsize = 15)
plt.ylabel('y', fontsize = 15)
plt.legend(fontsize = 12)
plt.axis('equal')
plt.grid(alpha = 0.3)
plt.axis([-2, 10, 1, 12])
plt.show()
     
```

```python
from sklearn import linear_model
clf = linear_model.LogisticRegression(solver = 'lbfgs')
clf.fit(np.vstack([X1,X2]), np.ravel(np.vstack([y1,y2])))
```

```python
w = np.zeros([3,1])

w0 = clf.intercept_[0]
w1 = clf.coef_[0,0]
w2 = clf.coef_[0,1]

xp = np.linspace(-2,10,100).reshape(-1,1)
yp = -w[1,0]/w[2,0]*xp - w[0,0]/w[2,0]
w12 = np.zeros([3,1])
plt.figure(figsize = (6, 4))
plt.plot(X1[:,0], X1[:,1], '.', markersize = 4, label = 'Class 1')
plt.plot(X2[:,0], X2[:,1], '.', markersize = 4, label = 'Class 2')
plt.plot(xp, yp, 'r--', linewidth = 2, label = '1 vs. 2')
plt.title('Multiclass Classification')
plt.axis('equal')
plt.grid(alpha = 0.3)
plt.axis([-2, 10, 0, 12])
plt.xlabel(r'x')
plt.ylabel(r'y')
plt.legend(loc = 1)
plt.show()
```
