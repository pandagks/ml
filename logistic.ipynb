```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

#odds 함수

p = np.linspace(0.01,0.99,100)

odds = p / (1-p)

plt.figure(figsize = (10,6))
plt.plot(p,odds, color = 'k')
plt.xlabel('Probability')
plt.ylabel('Odds ratio')
plt.title('Odds = p/(1-p)')
plt.show()
```

```python
#logit함수
p = np.linspace(0.01,0.99,100)

logit = np.log(p/(1-p))

plt.figure(figsize = (10,6))
plt.plot(p,logit,color='k')
plt.xlabel('Probabilty')
plt.ylabel('Logit')
plt.title('Logit = log(p/(1-p))')
plt.show()
```

```python
#sigmoid 함수

z = np.linspace(-10, 10 , 100)

sigmoid = 1 / (1 + np.exp(-z))

plt.figure(figsize = (10,6))
plt.plot(z,sigmoid,color = 'b')
plt.xlabel('z')
plt.ylabel('Probability')
plt.title('Sigmoid function')
plt.grid(alpha = 0.3)
plt.show()

def sigmoid_diff(z):
    sigmoid = 1 / (1+np.exp(-z))
    return sigmoid * (1 - sigmoid)

z = np.linspace(-10, 10 , 100)

plt.figure(figsize = (10,6))
plt.plot(z,sigmoid_diff(z),color = 'r')
plt.xlabel('z')
plt.ylabel('Probability')
plt.title('Derivative of the Sigmoid function')
plt.grid(alpha = 0.3)
plt.show()
```
