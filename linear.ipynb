```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 데이터 불러오기
x = pd.read_csv("C:\\Users\\panda\\Documents\\data0405.csv", usecols=['x']).values
y = pd.read_csv("C:\\Users\\panda\\Documents\\data0405.csv", usecols=['y']).values

# 초기 파라미터
w1 = 0
w0 = 0

lrate = 0.001
epochs = 500
n = float(len(x))

# 경사하강법 루프
for i in range(epochs):
    y_pred = w1 * x + w0
    
    # gradient 계산
    dw1 = (1/n) * np.sum((y_pred - y) * x)
    dw0 = (1/n) * np.sum(y_pred - y)
    
    # 파라미터 업데이트
    w1 = w1 - lrate * dw1
    w0 = w0 - lrate * dw0

print("w1:", w1)
print("w0:", w0)

xp = np.linspace(x.min() - 1, x.max() + 1, 100).reshape(-1, 1)
y_line = w1 * xp + w0

plt.figure(figsize=(6, 4))
plt.plot(x, y, 'bo', label='Data')
plt.plot(xp, y_line, 'r-', linewidth=2, label='Linear Regression')
plt.title('Gradient Descent: Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 데이터 불러오기
x = pd.read_csv("C:\\Users\\panda\\Documents\\data0405.csv", usecols=['x']).values
y = pd.read_csv("C:\\Users\\panda\\Documents\\data0405.csv", usecols=['y']).values

# 초기 파라미터
w2 = 0
w1 = 0
w0 = 0

lrate = 0.001
epochs = 500
n = float(len(x))

# 경사하강법 루프
for i in range(epochs):
    y_pred =w2*x**2 + w1 * x + w0
    
    # gradient 계산
    dw2 = (1/n) * np.sum((y_pred-y) * x**2)
    dw1 = (1/n) * np.sum((y_pred - y) * x)
    dw0 = (1/n) * np.sum(y_pred - y)
    
    # 파라미터 업데이트
    w2 = w2 -lrate * dw2
    w1 = w1 - lrate * dw1
    w0 = w0 - lrate * dw0

print("w2:",w2)
print("w1:", w1)
print("w0:", w0)

xp = np.linspace(x.min() - 1, x.max() + 1, 100).reshape(-1, 1)
y_line = w2 * xp**2 + w1 * xp + w0

plt.figure(figsize=(6, 4))
plt.plot(x, y, 'bo', label='Data')
plt.plot(xp, y_line, 'r-', linewidth=2, label='Linear Regression')
plt.title('Gradient Descent: Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x = pd.read_csv("C:\\Users\\panda\\Documents\\data0405.csv", usecols=['x']).values
y = pd.read_csv("C:\\Users\\panda\\Documents\\data0405.csv", usecols=['y']).values

w4 = 0
w3 = 0
w2 = 0
w1 = 0
w0 = 0

lrate = 0.001
epochs = 500
n = float(len(x))

for i in range(epochs):
    y_pred = w4* x**4 + w3 * x**3 + w2*x**2 + w1 * x + w0
    
    dw4 = (1/n) * np.sum((y_pred - y) * x**4)
    dw3 = (1/n) * np.sum((y_pred - y) * x**3)
    dw2 = (1/n) * np.sum((y_pred-y) * x**2)
    dw1 = (1/n) * np.sum((y_pred - y) * x)
    dw0 = (1/n) * np.sum(y_pred - y)
    
    w4 = w4 -lrate * dw4
    w3 = w3 -lrate * dw3
    w2 = w2 -lrate * dw2
    w1 = w1 - lrate * dw1
    w0 = w0 - lrate * dw0


print("w4:",w4)
print("w3:",w3)
print("w2:",w2)
print("w1:", w1)
print("w0:", w0)

xp = np.linspace(x.min() - 0.25, x.max() + 1, 100).reshape(-1, 1)
y_line = w4 * xp**4 + w3 * xp ** 3 + w2 * xp**2 + w1 * xp + w0

plt.figure(figsize=(6, 4))
plt.plot(x, y, 'bo', label='Data')
plt.plot(xp, y_line, 'r-', linewidth=2, label='Linear Regression')
plt.title('Gradient Descent: Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 데이터 불러오기
x = pd.read_csv("C:\\Users\\panda\\Documents\\data0405.csv", usecols=['x']).values
y = pd.read_csv("C:\\Users\\panda\\Documents\\data0405.csv", usecols=['y']).values

# 표준화
scaler_x = StandardScaler()
scaler_y = StandardScaler()
x_scaled = scaler_x.fit_transform(x)
y_scaled = scaler_y.fit_transform(y)

# 초기 가중치
w4 = 0
w3 = 0
w2 = 0
w1 = 0
w0 = 0

lrate = 0.001
epochs = 500
n = float(len(x_scaled))

# 경사하강법 학습
for i in range(epochs):
    y_pred = w4 * x_scaled**4 + w3 * x_scaled**3 + w2 * x_scaled**2 + w1 * x_scaled + w0

    dw4 = (1/n) * np.sum((y_pred - y_scaled) * x_scaled**4)
    dw3 = (1/n) * np.sum((y_pred - y_scaled) * x_scaled**3)
    dw2 = (1/n) * np.sum((y_pred - y_scaled) * x_scaled**2)
    dw1 = (1/n) * np.sum((y_pred - y_scaled) * x_scaled)
    dw0 = (1/n) * np.sum(y_pred - y_scaled)

    w4 -= lrate * dw4
    w3 -= lrate * dw3
    w2 -= lrate * dw2
    w1 -= lrate * dw1
    w0 -= lrate * dw0

# 결과 출력
print("w4:", w4)
print("w3:", w3)
print("w2:", w2)
print("w1:", w1)
print("w0:", w0)

# 예측 시각화
xp = np.linspace(x_scaled.min() - 1, x_scaled.max() + 1, 100).reshape(-1, 1)
y_line = w4 * xp**4 + w3 * xp**3 + w2 * xp**2 + w1 * xp + w0

# 원래 스케일로 복원
xp_original = scaler_x.inverse_transform(xp)
y_line_original = scaler_y.inverse_transform(y_line)

plt.figure(figsize=(6, 4))
plt.plot(x, y, 'bo', label='Data')
plt.plot(xp_original, y_line_original, 'r-', linewidth=2, label='4차 Polynomial Regression')
plt.title('Gradient Descent: 4th Degree Polynomial Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```
